```python
import numpy as np

def part(N, tmin, tmax):
    h = (tmax - tmin)/N
    
    t = np.zeros(N+1)
    for i in range(len(t)):
        t[i] = tmin + i*h
    
    return t, h

def solucao_edp_parabolica_progressiva(a, bottom_function, left_function, right_function, n, m, xmin, xmax, tmin, tmax, name=None, sol_exata=None):
    def f_1(x):
        f_1 = eval(left_function)
        return f_1
    
    def f_2(x):
        f_2 = eval(right_function)
        return f_2
    
    def f_3(x):
        f_3 = eval(bottom_function)
        return f_3
    
    x,h = part(n+1, xmin, xmax)
    t,k = part(m+1, tmin, tmax) 
    
    sigma = a*k/h**2
    
    # SOLUÇÃO INICIAL
    SOLUTION = np.zeros((m+2,n+2))

    SOLUTION[0:,0] = np.flip(f_1(t))
    SOLUTION[0:,n+1] = np.flip(f_2(t))
    SOLUTION[m+1,0:] = f_3(x)
    
    # print(SOLUTION)
    
    for i in reversed(range(m+1)): 
        for j in range(1, n+1): 
            SOLUTION[i][j] = SOLUTION[i+1][j] + sigma*(SOLUTION[i+1][j-1] - 2*SOLUTION[i+1][j] + SOLUTION[i+1][j+1])
            # print(f"[{i}, {j}] ::: {SOLUTION[i,j]} = {SOLUTION[i+1][j]} + {sigma}*({SOLUTION[i+1][j-1]} - {2*SOLUTION[i+1][j]} + {SOLUTION[i+1][j+1]})")
            
    import matplotlib.pyplot as plt

    # Iterate through each row of the matrix and plot
    for row in SOLUTION:
        plt.plot(row)
        
    # Show plot
    if name is not None:
        plt.savefig(f"{name}.pdf")
    plt.show()
        
    Xr, Tr = np.meshgrid(x,t)
    
    if sol_exata is not None:
        def u(x,t):
            u = eval(sol_exata)
            return u
                
        R = u(Xr, Tr)
        T = np.zeros(R.shape)

        for i in range(len(R)):
            T[i] = R[len(R) -1 - i]
        
        print("SOLUÇAO REAL ::: ")
        for i in range(len(T)):
            for j in range(len(T[0])):
                if T[i,j] < 0:
                    print(f"{T[i,j]:.5f} ", end=' ')
                else:
                    print(f" {T[i,j]:.5f} ", end=' ')
                
            print()
            
        e = abs(SOLUTION - T)
        print("ERRO ::: ")
        for i in range(len(T)):
            for j in range(len(T[0])):
                if e[i,j] < 0:
                    print(f"{e[i,j]:.5f} ", end=' ')
                else:
                    print(f" {e[i,j]:.5f} ", end=' ')
                
            print()
            
    return SOLUTION

# homogenea
def solucao_edp_parabolica_regressiva(a, bottom_function, left_function, right_function, n, m, xmin, xmax, tmin, tmax, name=None, sol_exata=None):
    def f_1(x):
        f_1 = eval(left_function)
        return f_1
    
    def f_2(x):
        f_2 = eval(right_function)
        return f_2
    
    def f_3(x):
        f_3 = eval(bottom_function)
        return f_3
    
    x,h = part(n+1, xmin, xmax)
    t,k = part(m+1, tmin, tmax) 
    
    sigma = a*k/h**2
    
    # SOLUÇÃO INICIAL
    SOLUTION = np.zeros((m+2,n+2))

    SOLUTION[m+1,0:] = f_3(x)
    SOLUTION[0:,0] = np.flip(f_1(t))
    SOLUTION[0:,n+1] = np.flip(f_2(t))
    
    A = np.zeros((n,n))
    for i in range (len(A)):
        A[i,i] = 1 + 2*sigma
        if i-1 >= 0:
            A[i,i-1] = -sigma
            
        if i+1 < len(A):
            A[i,i+1] = -sigma
            
            
    for l in reversed(range(m+1)):                
        b = np.zeros(n)
        for i in range(len(b)):
            b[i] = SOLUTION[l+1][i+1]
            
        SOLUTION[l,1:n+1] = np.linalg.solve(A,b)
        
    import matplotlib.pyplot as plt

    # Iterate through each row of the matrix and plot
    for row in SOLUTION:
        plt.plot(row)
        
    # Show plot
    if name is not None:
        plt.savefig(f"{name}.pdf")
    plt.show()
        
    Xr, Tr = np.meshgrid(x,t)
    
    if sol_exata is not None:
        def u(x,t):
            u = eval(sol_exata)
            return u
                
        R = u(Xr, Tr)
        T = np.zeros(R.shape)

        for i in range(len(R)):
            T[i] = R[len(R) -1 - i]
        
        print("SOLUÇAO REAL ::: ")
        for i in range(len(T)):
            for j in range(len(T[0])):
                if T[i,j] < 0:
                    print(f"{T[i,j]:.5f} ", end=' ')
                else:
                    print(f" {T[i,j]:.5f} ", end=' ')
                
            print()
            
        e = abs(SOLUTION - T)
        print("ERRO ::: ")
        for i in range(len(T)):
            for j in range(len(T[0])):
                if e[i,j] < 0:
                    print(f"{e[i,j]:.5f} ", end=' ')
                else:
                    print(f" {e[i,j]:.5f} ", end=' ')
                
            print()
        
    return SOLUTION
    
```
